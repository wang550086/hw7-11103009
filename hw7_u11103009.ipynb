```python
import numpy as np
```

```python
lat = 48.1195; lon = 12.1878;
ele = 487.9; go = 980717.39; ge = 978031.85;
np.pi
fac = np.pi/180.0;
```

```python
gt = ge * ( 1 + 0.005278895 * np.sin(lat*fac)**2 + 0.000023462 * np.sin(lat*fac)**4);
print(f"theoretical gravtity : {gt} mGal")
FAC = ele*0.308
print(f"Free-air correction : {FAC} mGal")
BC = ele*0.112
print(f"Bouguer correction : {BC} mGal")
gfa = go-gt+FAC
print(f"Free-air gravity anomaly : {gfa} mGal")
gbc = gfa - BC
print(f"Bouguer gravity anomaly : {gbc} mGal")
```

```python
BCs = -0.0687*ele
FACs = 0*0.308
gfas= go - gt + FACs
print(f"Free-air gravity anomaly on the ocean : {gfas} mGal")
gBCs = gfas - BCs
print(f"Bouguer gravity anomaly on the ocean : {gBCs} mGal")
```
